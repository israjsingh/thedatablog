```python
import numpy as np

class NeuralNetwork:
    
    def __init__(self, design, input, weights, thresholds):
        self.number_of_layers = np.size(design)
        self.input_val = input
        self.design = design
        self.threshold_inputs = thresholds
        self.weight_inputs = weights
        
        
    def __assign_weights(self):
        
        """Note that we are ommitting the last layer i.e. output kayer from 
        weights assignment. Output ayer doesnot contain weights as it doesnot act as an input """
        
        print "======Assigning Weights======"
        self.weights = {}
        for index, perceptron_count in enumerate(self.design[:-1]):
            i = 0
            while i < perceptron_count:
                self.weights["w{}{}".format(index,i)] = self.weight_inputs[index][i]
                print "w"+str(index)+str(i) + "=" + str(self.weights["w{}{}".format(index,i)])
                i = i+1
                
    def __assign_thresholds(self):
        
        """Note that we are not assigning threshold for input layer, 
        as thresholds are needed only for layers that act as output layers"""
        
        print "======Assigning Thresholds======"
        self.thresholds = {}
        for index, perceptron_count in enumerate(self.design[1:]):
            i = 0
            while i < perceptron_count:
                self.thresholds["t{}{}".format(index+1,i)] = self.threshold_inputs[index][i]
                print "t"+str(index+1)+str(i) + "=" + str(self.thresholds["t{}{}".format(index+1,i)])
                i = i+1
     
    def __assign_input(self):
        
        print "======Assigning Inputs======"
        self.inputs = {}
        for index,val in enumerate(self.input_val) :
            self.inputs["x0{}".format(index)]=val
            print "x0"+str(index) + "=" + str(self.inputs["x0{}".format(index)])
            
    def __calculate(self):
        output = {}
        for i in xrange(0,self.design[0]):
            output['o1{}'.format(i)] = self.inputs['x0{}'.format(i)]*self.weights['w0{}'.format(i)]
        activation_value = sum(output.values())
        
        for index,val in enumerate(self.design[1:-1]):
            output = {}
            for j in xrange(0,val):
                if self.thresholds['t{}{}'.format(index+1,j)]<activation_value:
                    self.inputs["x{}{}".format(index+1,j)]=1
                else :
                    self.inputs["x{}{}".format(index+1,j)]=0
                output['o{}{}'.format(index+1,j)] = self.inputs["x{}{}".format(index+1,j)]*self.weights["w{}{}".format(index+1,j)]
            activation_value = sum(output.values())
        if activation_value>self.threshold_inputs[-1][0]:
            self.final_output = 1
        else :
            self.final_output = 0
          
                  
    def run(self):
        self.__assign_input()
        self.__assign_weights()
        self.__assign_thresholds()
        self.__calculate()
        return self.final_output
    
"""
Assuming we are creating 3 layered neural networl where the first layer is input layer, 
follwed by a hidden layer and then the final output layer. 
In this example, the input layer jas 2 inputs, hidden layer has 3 
perceptrons and out layer has 1 perceptron 
"""

output = NeuralNetwork(design=[2,3,2,1],
                        input=[1,1], 
                        weights=[[2,1],[2,1,3],[1,2]], 
                        thresholds=[[5,2,1],[2,3],[3]]).run()
print "output = "+str(output)
        
```
